## 2. Writing tests

To do the tests we will make assertions with the graphs as geodataframes:

```python
import momepy as mm
```

```python
tol=30
for method in ['spider', 'euclidean', 'extend']:
    graph_simplified = consolidate_intersections(graph, tolerance=tol, rebuild_graph=True, rebuild_edges_method=method)
    nodes_simplified, edges_simplified = mm.nx_to_gdf(graph_simplified)
    
    assert len(nodes_simplified) == 39
    assert len(edges_simplified) == 66
    
    if method != 'euclidean':
        assert edges_simplified.length.min() >= tol
        
```
