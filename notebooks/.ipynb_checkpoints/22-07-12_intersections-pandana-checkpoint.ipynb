# Consolidating Intersections with Pandana and Coins

```python
dist_threshold=25 #parameter to consolidate intersections. This has different meanings depending on whether we use OSMnx or Pandana, will be clarified
```

## 1. Collecting Data

```python
point = 40.432996, -3.691383 #same as Greg
dist = 500
```

```python
graph = ox.graph_from_point(point, dist=dist,
                            network_type='drive', simplify=True)
graph_und = ox.get_undirected(graph)
graph_proj_und = ox.project_graph(graph_und)
graph_proj = ox.project_graph(graph)
nodes, edges = ox.graph_to_gdfs(graph_und,
                                nodes=True, edges=True,
                                node_geometry=True, fill_edge_geometry=True)
```

```python
fig, ax = ox.plot_graph(graph_und, figsize=(15,15), node_color='k', node_size=40, bgcolor='white')
plt.show()
```

## 2. Pandana

```python
edges_flat = edges.reset_index()
pandana_graph=pandana.Network(nodes['x'], nodes['y'],
                              edges_flat['u'], edges_flat['v'],
                              edges_flat[['length']])
pandana_graph.precompute(dist_threshold+1) #Essentially, this will cache all distance computations for our distance threshold
```

If we were to find the nodes tha make clusters we could find the nodes with "high-connectivity": connected to at least one node within our threshold

```python
low_connectivity_nodes = pandana_graph.low_connectivity_nodes(dist_threshold, 2, 'length')
high_connectivity_nodes = pandana_graph.node_ids.difference(low_connectivity_nodes)
```

```python
fig, ax = ox.plot_graph(graph_und,
                        node_color=['red' if node in high_connectivity_nodes else 'white' for node in pandana_graph.node_ids.to_list()], node_size=50, node_zorder=1,
                        figsize=(20,20))
```

This just finds the nodes we want to to merge without assigning them to the specific cluster. We can identify clusters using the POI method.

```python
high_connectivity_nodes_df = nodes[['x', 'y']].loc[high_connectivity_nodes]
pandana_graph.set_pois(category='intersection', maxdist=dist_threshold, maxitems=100,
                       x_col=high_connectivity_nodes_df.x, y_col=high_connectivity_nodes_df.y)
closest_nodes_df = pandana_graph.nearest_pois(dist_threshold, 'intersection',
                                              num_pois=100,
                                              imp_name='length', include_poi_ids=True)
```

```python
closest_nodes_df.head()
```

This returns a dataframe containing all nodes in the graph. No additional computation was made since we precomputed distances. However, there are many "nan" values which we now must clean:

```python
high_connectivity_closest_nodes_df = closest_nodes_df.loc[high_connectivity_nodes, 'poi1':].dropna(axis=1, how='all')
high_connectivity_closest_nodes_dict = high_connectivity_closest_nodes_df.transpose().to_dict(orient='list')
high_connectivity_closest_nodes_dict_clean = {}
for node in high_connectivity_closest_nodes_dict:
    items_cleaned = [int(item) for item in high_connectivity_closest_nodes_dict[node] if not np.isnan(item)]
    high_connectivity_closest_nodes_dict_clean[node] = items_cleaned
```
